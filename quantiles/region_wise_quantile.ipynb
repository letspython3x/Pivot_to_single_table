```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('records_50000.csv')
# df = df[["gdenr", "dist_see", "egid"]]
df['points'] = np.nan
df.head()
```

```python
def get_points_on_quants(quants, val):    
    if quants[0.7] <= val < quants[0.8]:
        return 8
    elif quants[0.8] <= val < quants[0.9]:
        return 9    
    elif quants[0.9] <= val:
        return 10
    else:
        return np.nan
```

```python
groups = df.groupby(["gdenr"]).groups
for region in groups:
    data = df[df.gdenr == region]
    region_quants = data["dist_see"].dropna().quantile([0.7, 0.8, 0.9])    
    df.loc[df.gdenr == region, 'points'] = data['dist_see'].apply(lambda row: get_points_on_quants(region_quants, row))
```
